```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Heart (1) .csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df["Age"], y=df["MaxHR"], hue=df["AHD"], alpha=0.7, edgecolor="black", s=80)
plt.title("Age vs Max Heart Rate", fontsize=14, fontweight='bold')
plt.xlabel("Age", fontsize=12)
plt.ylabel("Maximum Heart Rate", fontsize=12)
plt.legend(title="Heart Disease", title_fontsize='13', fontsize='11')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x=df["ChestPain"], order=df["ChestPain"].value_counts().index)
plt.title("Chest Pain Types Distribution")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=df["AHD"], y=df["Chol"])
plt.title("Cholesterol Levels by Heart Disease")
plt.show()
```

```python
plt.figure(figsize=(6, 6))
heart_disease_counts = df["AHD"].value_counts()
plt.pie(heart_disease_counts, labels=heart_disease_counts.index, autopct="%1.1f%%", colors=["lightblue", "salmon"])
plt.title("Proportion of Heart Disease Cases")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.lineplot(x=df["Age"], y=df["Chol"], marker="o")
plt.title("Cholesterol Levels with Age")
plt.show()
```
