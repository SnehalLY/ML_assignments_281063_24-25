```python
#Indexing and selecting data, sort data
df.loc[0:3, ['CustomerID', 'Genre']]
df.iloc[0:3, 0:2]
```

```python
#Sort data by a specific column
sorted_df = df.sort_values(by='Genre', ascending=False)  
sorted_df
```

```python
#Sort data by multiple columns
sorted_df_multi = df.sort_values(by=['Age', 'Annual Income (k$)'], ascending=[True, False])  
sorted_df_multi
```

```python
#checking datatypes of each column
df.dtypes
```

```python
columns = df.columns
columns
```

```python
df.columns = df.columns.str.strip()

# Checking for leading/trailing spaces in string columns
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].str.strip()
df
```

```python
#Counting Unique Values for Each Column
unique_counts = df.nunique()
unique_counts
```

```python
#Converting Variable Data Types
df['Age'] = df['Age'].astype(float)
df['Age'].dtypes
```

```python
#checking data type after conversion
df.dtypes
```

```python
#Renaming Columns
df.rename(columns={'Annual Income (k$)': 'Annual Income'}, inplace=True)
df
```
